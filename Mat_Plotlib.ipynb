```python
 # This program displays a simple line graph.
import matplotlib.pyplot as plt
def main():
# Create lists with the X,Y coordinates of each data point
  x_coords = [0, 1, 2, 3, 4]
  y_coords = [0, 3, 1, 5, 2]
# Build the line graph.
plt.plot(x_coords, y_coords, marker='o')
# Add a title
plt.title('Sales by Year')
# Add labels to the axes
plt.xlabel('Year')
plt.ylabel('Sales')
```

```python
plt.xticks([0,1,2,3,4],['2016', '2017', '2018', '2019', '2020'])
plt.yticks([0, 1, 2, 3, 4, 5], ['$0m', '$1m', '$2m', '$3m', '$4m', '$5m'])
# Add a grid.
plt.grid(True)
# Display the line graph.
plt.show()
# Call the main function.
if __name__ == "__main__":
  main()
```

```python
#Plotting a bar chart
left_edges = [0, 10, 20, 30, 40]
heights = [100, 200, 300, 400, 500]
plt.bar(left_edges, heights)
plt.bar(left_edges, heights, color=('r', 'g', 'b', 'm', 'k'))
plt.show()
left_edges = [0, 10, 20, 30, 40]
heights = [100, 200, 300, 400, 500]
bar_width = 5
#plt.bar(left_edges, heights, bar_width)
plt.bar(left_edges, heights, bar_width, color=('r', 'g', 'b', 'm', 'k'))
plt.show()
```

```python
values = [20, 60, 80, 40]
plt.pie(values)
plt.pie(values, colors=('g', 'b', 'm', 'k'))
plt.show()
sales = [100, 400, 300, 600]
slice_labels = ['1st Qtr', '2nd Qtr', '3rd Qtr', '4th Qtr']
plt.pie(sales, labels=slice_labels)
plt.pie(values, colors=('g', 'b', 'm', 'k'))
plt.title('Sales by Quarter')
plt.show()
```
